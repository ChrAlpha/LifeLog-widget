```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
full = sum([1 / i for i in range(1, n + 1)])

values = [sum([1 / j for j in range(1, i+1)]) for i in range(1, n + 1)]
proportions = [val / full for val in values]

plt.figure(figsize=(10, 6))
bars = plt.bar(range(1, n + 1), proportions)
plt.xlabel("i")
plt.ylabel("Proportion of first 1/i sum to total")
plt.title("Proportion of first 1/i sum to the total sum")
plt.grid(axis="y", alpha=0.3)

plt.plot(range(1, n + 1), np.log(np.array(range(1, n + 1)) + 1) / np.log(n), color="red", label="log(i+1)")
plt.legend()

for i, bar in enumerate(bars[:10]):
    plt.text(
        i + 1,
        bar.get_height() + 0.01,
        f"{proportions[i]:.1%}",
        ha="center",
        va="bottom",
        fontsize=8,
    )

plt.tight_layout()
plt.show()
```
